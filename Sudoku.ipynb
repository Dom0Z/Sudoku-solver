```
def printBoard(b):
  for i in range(len(b)):
    if i % 3 ==0 and i != 0:
        print(" - - - - - - - - - - - " )
    for j in range (len(b[0])):
        if j % 3 == 0 and j != 0:
          print("|",end="")
        if j == 8:
          print(b[i][j])
        else:
          print(str(b[i][j]) + " ",end="")
  
```

```
def findEmpty(b):
  for i in range (len (b)): #row
    for j in range(len(b[0])): #col
      if(b[i][j] == 0):
        return (i, j) 
  return None
```

```
findEmpty(SudokuBoard)
```

```
def Validity(b, num, pos):
  for i in range(len(b[0])):
    # check row
    if b[pos[0]] [i] == num and pos[1] != i:
      return False
    #check col
  for i in range(len(b)):
    if b[i][pos[1]] == num and pos[0] != i:
      return False
    #check box
  boxX = pos[1] // 3
  boxY = pos[0] // 3
  for i in range(boxY*3, boxY*3 + 3):
      for j in range(boxX * 3, boxX*3 + 3):
          if b[i][j] == num and (i,j) != pos:
              return False
  return True
```

```
def solve(b):
  empty = findEmpty(b)
  if not empty:
    return True
    
  else:
    row, col = empty
  for i in range (1,10):
    
    if Validity(b,i,(row,col)):
      b[row][col] = i
      
      if solve(b):
        return True
        #this is the actual back tracking as if this didnt work we set it back to zero and try again
      b[row][col] = 0
  return False
```

```
solve(SudokuBoard)
printBoard(SudokuBoard)
```
